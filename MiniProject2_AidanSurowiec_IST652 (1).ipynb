```python
# The ntweet list has a limit of 200, had to use the cursor method to get our desired 600 tweets 
# Create a pandas data frame to get each text value from each tweet in out the netflix user timeline - label the column as Tweets 
netflix_df = pd.DataFrame(data=[tweet.text for tweet in tweepy.Cursor(api.user_timeline, id="netflix").items(600)], columns=['Tweets'])

# Show the first 10 instances in the data frame to make sure it turned out as intended 
netflix_df.head(10)
```

```python
#Import numpy so we can add other necessary information to our data frame - have data correlate as intended 
import numpy as np
```

```python
#Need the Cursor iteration in each np.array so that we can build a data frame with 600 tweets
#Could to "for tweet in ntweets", but would be limited to 200 tweets
#Want several hundred results for deeper analysis 
#User the same calls as listed above - .id, .created_at etc
#Add each as a column in the netflix df 

# Add length variable to the data frame - measures the number of total characters in the text 
netflix_df['len']  = np.array([len(tweet.text) for tweet in tweepy.Cursor(api.user_timeline, id="netflix").items(600)])
# Add ID number to keep track of who made each tweet 
netflix_df['ID']   = np.array([tweet.id for tweet in tweepy.Cursor(api.user_timeline, id="netflix").items(600)])
# Add a Date column to insert the created at values - will be necessary for breakdown of tweets per hour 
netflix_df['Date'] = np.array([tweet.created_at for tweet in tweepy.Cursor(api.user_timeline, id="netflix").items(600)])
# Add source column to understand what medium the tweet was posted from 
netflix_df['Source'] = np.array([tweet.source for tweet in tweepy.Cursor(api.user_timeline, id="netflix").items(600)])
# Add favorites column to keep track of the favorite count for each tweet 
netflix_df['Favs']  = np.array([tweet.favorite_count for tweet in tweepy.Cursor(api.user_timeline, id="netflix").items(600)])
# Add RTs column to keep track of the retweet count for each tweet 
netflix_df['RTs']    = np.array([tweet.retweet_count for tweet in tweepy.Cursor(api.user_timeline, id="netflix").items(600)])
```

```python
# Show the head of the data frame after adding these new columns 
netflix_df.head(10)
```

```python
#Import datetime so that we can do analysis on when tweets were created 
from datetime import datetime
```

```python
#Create a convert date function to take the Date data -and convert to datetime object 
#Need to nest the date in str() to make it a string - right now it is a timestamp 
def convert_date(date):
    """
    Takes date as string
    Returns date as datetime.datetime object
    """
    return datetime.strptime(str(date), '%Y-%m-%d %H:%M:%S')
```

```python
#Apply function to every tweet in the date column in the netflix dataframe 
netflix_df['Date'] = netflix_df['Date'].apply(lambda x: convert_date(x))
```

```python
#Apply function to every tweet in the date column in the hulu dataframe 
hulu_df['Date'] = hulu_df['Date'].apply(lambda x: convert_date(x))
```

```python
#Create bins - iterate by 1 from the range 0-24
#Must make the range 25, or times in between 23-24 hours will be left out 
bins = [x for x in range (0,25,1)]
bins


# Add these new bins to our netflix data frame 
#In the data frame - use cut function on Date to only target the hour 
#create new column called Time Bin - the hour will categorize which bin this is in 
netflix_df['Time Bin'] = pd.cut(netflix_df.Date.dt.hour, bins, right=False)

#Repeat for hulu data frame 
hulu_df['Time Bin'] = pd.cut(hulu_df.Date.dt.hour, bins, right=False)
```

```python
#Show the first 10 rows of the netflix data frame - ensure that the code works 
netflix_df.head(10)
```

```python
#Show the first 10 rows of the hulu data frame - ensure that the code works 
hulu_df.head(10)
```

```python
#Group the dataframe by the Time Bins, and then return the count of occurrences for each time bin
#This will show how many tweets were created within that binned hour 
netflixph = netflix_df.groupby(['Time Bin'])['Time Bin'].count()

#Return to console to ensure the frequency was recorded correctly 
netflixph
```

```python
#Put this into a data frame 
nfph = pd.DataFrame(netflixph)

#Rename the time bin column to Netflix Tweets Per Hour 
#This will make analysis easier in the newly exported file
nfph = nfph.rename(columns={"Time Bin": "Netflix Tweets Per Hour"})
```

```python
#Repeat the process for Hulu
huluph = hulu_df.groupby(['Time Bin'])['Time Bin'].count()
print(huluph)
```

```python
#Put the new grouped data into a dataframe 
hfph = pd.DataFrame(huluph)

#Rename the time bin column to make analysis easy 
hfph = hfph.rename(columns={"Time Bin": "Hulu Tweets Per Hour"})
```

```python
#Create a variable that has both data frames stored 
perhour_dfs = [nfph, hfph]
#concatenate the data frames on a single axis, inner join so that they are compared at each bin
perhourcompare = pd.concat(perhour_dfs, axis=1, join='inner')
#Create output file number 2 
perhourcompare.to_csv('perhourcompare.csv')
```

```python
#Create a new groupby, group the data frame by the source data
#Use .agg to use multiple aggregate functions in this one groupby formula
#Want the tweet count for each source, and the Fav/RT statistics for each
netflixs = netflix_df.groupby(['Source']).agg({'RTs': ['mean', 'min', 'max'], 'Favs': ['mean', 'min', 'max'], 'Source': ['count']})
#Return output to console 
netflixs
```

```python
#Repeat this process for Hulu
hulus = hulu_df.groupby(['Source']).agg({'RTs': ['mean', 'min', 'max'], 'Favs': ['mean', 'min', 'max'], 'Source': ['count']})
hulus
```

```python
#Create a comparison variable that will compare both source stats for Hulu and netflix
#Concatenate the two data frames on the 0 Axis, inner join so that the RTs, Favs, and Source are not duplicated 
#List the keys to be Netflix and Hulu so that identifying each data set is easier 
sourcecompare = pd.concat([netflixs, hulus], axis=0, join="inner", keys=['Netflix', 'Hulu'])
#Create the file output 3 for analysis 
sourcecompare.to_csv('sourcecompare.csv')
```
